### **Завдання 1 (2 бали)**

Розрахуйте коефієнти розкладу біному Ньютона за допомогою генератора (https://uk.wikipedia.org/wiki/%D0%91%D1%96%D0%BD%D0%BE%D0%BC_%D0%9D%D1%8C%D1%8E%D1%82%D0%BE%D0%BD%D0%B0).

1. Степінь многочлена користувач має вводити з клавіатури (невід'ємне ціле число).

2. Програма має підраховувати біноміальні коефіцієнти для біномів усіх степенів, починаючи від 0 і закінчуючи тим, який ввів користувач.

3. Коефіцієнти мають виводитись у вигляді трикутника Паскаля.

Наприклад, якщо користувач введе число 5, програма має вивести результат:

1

1 1

1 2 1

1 3 3 1

1 4 6 4 1

1 5 10 10 5 1


```
import math
def coefficient(n, k):
    a = math.factorial(n)
    b = math.factorial(k)
    c = math.factorial(n-k)
    return a//(b*c)
def input_number(message):
    while True:
        try:
            num = int(input(message))
            if num < 0:
                print("Число повинно бути цілим і додатним")
            else:
                return num
        except ValueError:
            print("Ви ввели недопустиме значення")
n = input_number("Введіть число: ")
for i in range(0, n+1):
    for k in range(0, i+1):
        print(coefficient(i, k), end=" ")
        k = k+1
    print()
```

6. Замість функції з п. 4 створіть генератор, який буде визначати послідовність книжкових сторінок для кожного зошита. Ці рядки можна потім вивести або у циклі, або за допомогою функції next(). Функція-декоратор має працювати з цим генератором так само, як і зі звичайною функцією.

```
while True:
    try:
        book_pages = int(input("Введіть кількість сторінок книги: "))
        if book_pages not in range(0, 1281):
            print("Число не повинно перевищувати 1280!")
        else:
            break
    except ValueError:
        print("Ви ввели недоступне значення")
        continue

while True:
    try:
        notebook_pages = int(input("Введіть кількість сторінок книги (допустимі значення - 16, 24, 32):"))
        if notebook_pages == 16 or notebook_pages == 24 or notebook_pages == 32:
            break
        else:
            print("Ваше число повинно бути 16, 24 або 32")
    except ValueError:
        print("Ви ввели недоступне значення")
        continue

while True:
    try:
        if book_pages % notebook_pages == 0:
            break
        else:
            print("Загальна кількість сторінок у книзі має бути кратною кількості сторінок у зошиті!")
            book_pages = int(input("Введіть кількість сторінок книги знову: "))
    except ValueError:
        print("Ви ввели недоступне значення")
        continue

while True:
    try:
        parameter = int(input("Введіть 0 або 1: "))
        if parameter == 0 or parameter == 1:
            break
        else:
            print("Ви можете вводити тільки 0 або 1")
    except ValueError:
        print("Ви ввели недоступне значення")
        continue

notebook_count = book_pages // notebook_pages

def decorator(param):
    def logger(func):
        def inner(books_pages, notebooks_pages):
            some_list = func(books_pages, notebooks_pages)
            if param:
                print(some_list)
                print("Кількість зошитів = ", notebook_count)
            else:
                for element in some_list:
                    for n in element:
                        print(n, end=' ')
                    print()
                print("Кількість зошитів = ", notebook_count)
        return inner
    return logger


@decorator(parameter)
def function(books_pages, notebooks_pages):
    some_list = []
    for i in range(1, notebook_count + 1):
        note_pages = []
        for j in range(0, notebooks_pages // 4):
            note_pages.append(notebook_pages * i - j * 2)
            note_pages.append(notebook_pages * i - notebook_pages + 1 + j * 2)
            note_pages.append(notebook_pages * i - notebook_pages + 2 + j * 2)
            note_pages.append(notebook_pages * i - 1 - j * 2)
        some_list.append(note_pages)
    return some_list

function(book_pages, notebook_pages)
```
